```python
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter
# Load the dataset
churn_data = pd.read_csv("Telco-Customer-Churn.csv") # update according to where the data is saved
# Create event indicator (1 = churned, 0 = active)
churn_data['event_observed'] = churn_data['Churn'].map({'Yes': 1, 'No': 0})
# Drop rows with missing 'tenure' values
churn_data = churn_data.dropna(subset=['tenure'])
# Initialize and fit the Kaplan-Meier estimator
kmf = KaplanMeierFitter()
kmf.fit(durations=churn_data['tenure'], event_observed=churn_data['event_observed'], label="Customer Survival")
# Plot the survival function
kmf.plot_survival_function()
plt.title("Kaplan-Meier Survival Curve - Telco Customers")
plt.xlabel("Tenure (Months)")
plt.ylabel("Probability of Retention")
plt.grid(True)
plt.show()
```

```python
from lifelines.datasets import load_dd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt
# Load regime duration dataset
regime_duration_dataset = load_dd()
# Initialize Kaplan-Meier fitter
kmf = KaplanMeierFitter()
# Fit on full dataset
kmf.fit(durations=regime_duration_dataset['duration'],
event_observed=regime_duration_dataset['observed'])
# Plot survival function
plt.figure(figsize=(10, 6))
kmf.plot_survival_function()
plt.title("Kaplan-Meier Survival Function: All Regimes")
plt.xlabel("Duration (Years)")
plt.ylabel("Survival Probability")
plt.grid(True)
plt.show()
```
