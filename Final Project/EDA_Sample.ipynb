Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sample_2019-Oct-Nov.csv')

print(df.head())

print(df.info())

print(df.shape)

# basic statistics
print(df.describe())
```

```python
# count of distinct values
print('event_type:', df['event_type'].nunique())
print('product_id:', df['product_id'].nunique())
print('category_id:', df['category_id'].nunique())
print('category_code:', df['category_code'].nunique())
print('brand:', df['brand'].nunique())
print('user_id:', df['user_id'].nunique())
```

```python
# plot frequency event_type
print(df['event_type'].value_counts().plot(kind='bar'))
plt.title('Frequency of event_type')
```

```python
# plot top 10 in frequency for brand
print(df['brand'].value_counts().head(10).plot(kind='bar'))
plt.title('Top 10 brand in frequency')
plt.xlabel('brand')
plt.ylabel('Frequency')
```

```python
# plot top 10 in frequency for category_code
print(df['category_code'].value_counts().head(10).plot(kind='bar'))
plt.title('Top 10 category_code in frequency')
plt.xlabel('category_code')
plt.ylabel('Frequency')
```
